```
'''1. Signal Classification Using Machine Learning
Problem Statement: You have a dataset of electrical signals, where each signal
represents either a &quot;sine wave&quot; or a &quot;square wave.&quot; The dataset contains the following
features for each signal: - Amplitude (in volts) , Frequency (in Hz) , Duration (in
seconds) , Peak-to-Peak Voltage (in volts) .
Your task is to develop a machine learning model that can classify each signal as
either &quot;Sine&quot; or &quot;Square&quot; based on these features using KNN. Encode labels (Sine →
0, Square → 1)'''
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from google.colab import files
import pandas as pd
uploaded=files.upload()
df=pd.read_csv('wave - Sheet1.csv')
print(df)
x=df['amplitude']
y=df['duration']
classes=df['label'].map({'Sine':0,'Square':1})
data=list(zip(x,y))
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(data,classes)
nx=1.4
ny=0.54
np=[(nx,ny)]
prediction=knn.predict(np)
plt.scatter(x+[nx],y+[ny],c=classes+prediction[0])
plt.xlabel("Amplitude")
plt.ylabel("Duration")
plt.title("Signal classification")
plt.show()
```

```
'''2. Write a program to classify a component is defective or not based on Voltage and
Current using KNN. Encode labels (Working → 1, Defective → 0)'''
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from google.colab import files
import pandas as pd
uploaded=files.upload()
df=pd.read_csv('defective - Sheet1.csv')
print(df)
x=df['Voltage']
y=df['Current']
classes=df['status'].map({'working':1,'defective':0})
data=list(zip(x,y))
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(data,classes)
nx=4
ny=0.55
np=[(nx,ny)]
prediction=knn.predict(np)
plt.scatter(x+[nx],y+[ny],c=classes + [prediction[0]])

plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("Defective component")
plt.show()
```

```
'''3. Write a program to Predict Final Exam Scores based on features like study hours
and Final exam score using Linear Regression.'''
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('Exam - Sheet1.csv')
df.head()
x=df['Study hours']
y=df['Final exam score']
slope,intercept,r,s,std_err=stats.linregress(x,y)
def func(x):
  return slope*x+intercept
m=list(map(func,x))
plt.scatter(x,y)
plt.plot(x,m)
plt.xlabel('Study hours')
plt.ylabel('Final exam score')
plt.title('prediction of exam score')
plt.show()
```

```
'''4. Write a program to build a machine learning linear regression model to predict
the price of a house based on its features. You are given a dataset containing
information of Year and price.'''
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('Price - Sheet1.csv')
print(df)
x=df['Year']
y=df['Price']
slope,intercept,r,s,std_err=stats.linregress(x,y)
def func(x):
  return slope*x+intercept
m=list(map(func,x))
plt.scatter(x,y)
plt.plot(x,m)
plt.xlabel('Year')
plt.ylabel('Price')
plt.title("Prediction of price")
plt.show()
```

```
'''5. Write a program to develop a decision tree classifier to predict signal quality (Good or
Bad) based on transmitter, signal strength, frequency.'''
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('Signal - Sheet1.csv')
df.head()
d={'T1':0,'T2':1,"T3":2}
df['Transmitter']=df['Transmitter'].map(d)
d={'Good':1,'Bad':0}
df['Signal quality']=df['Signal quality'].map(d)
features=['Transmitter','Signal strength','Frequency']
x=df[features]
y=df['Signal quality']
dtree=DecisionTreeClassifier()
dtree=dtree.fit(x,y)
tree.plot_tree(dtree,feature_names=features)
plt.xlabel("Transmitter")
plt.ylabel("Signal quality")
plt.title("Signal quality analyser")
plt.show()
```

```
'''6. Write a program to Generate a synthetic dataset representing electronic components based on their
voltage and current characteristics using Kmeans Clustering.
The dataset contains two clusters based on Voltage and Current.
The dataset contains two clusters based on Voltage and Current.
 Cluster 0 corresponds to low voltage and current values.
 Cluster 1 corresponds to higher voltage and current values.'''
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from google.colab import files
import pandas as pd
uploaded=files.upload()
df=pd.read_csv('VC - Sheet1.csv')
print(df)
x=df['Voltage']
y=df['Current']
d=list(zip(x,y))
kmeans=KMeans(n_clusters=2)
kmeans.fit(d)
plt.scatter(x,y,c=kmeans.labels_)
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("Electronic component")
plt.show()

```
