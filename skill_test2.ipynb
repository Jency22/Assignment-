```python
'''2.Write a Python script using pandas for Student Performance Analysis:
Problem Statement:
Perform the following tasks:
Display the summary statistics for each subject,
Calculate the average score for each student,
Identify students who scored below 60 in more than two subjects.'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('student_score - Sheet1.csv')
print(df)
a=df[['Maths','Physics','Chemistry']].describe()
print("Statistics for each subject:",a)
df['Average']=df[['Maths','Physics','Chemistry']].mean(axis=1)
print(df[['Student ID','Average']])
b=df[['Maths','Physics','Chemistry']]<60
df['b']=b.sum(axis=1)>2
print("\nStudents below 60 marks in more than 2 subjects:")
count=0
for k in range(len(df)):
  if df['b'][k]==True:
    print(df['Student ID'][k])
    count=1
if count==0:
  print("No student scored below 60 in morethan 2 subject")



```

```python
'''3.Write a Python script for Flight Data Analysis:
Problem Statement:
You have a CSV file named 'flight_data.csv' containing flight information such as origin, destination, and departure time.
Perform the following tasks:
Display the summary statistics for departure delays,
Calculate the average delay for each airline,
Identify the most common departure and arrival destinations,
Visualize the distribution of departure delays using a box plot.'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('flight - Sheet1.csv')
print(df)
a=df['Dep_delay'].describe()
print("Statistics for departure delays:",a)
df['Avg_delay']=df.groupby('Airline')['Dep_delay'].transform('mean')
print(df[['Airline','Dep_delay','Avg_delay']])
b=df.groupby('Airline').agg({'Dep_delay':'mean'})
print("\nAverage delay for each airline:",b)
c=df['Origin'].value_counts().idxmax()
d=df['Destination'].value_counts().idxmax()
print("\nMost common departure destination:",c)
print("Most common arrival destination:",d)
sns.boxplot(x='Dep_delay',data=df)
plt.title('Distribution of Departure Delays')
plt.show()




```

```python
'''5.Write a program to simulate and analyze daily stock prices over one year and generate
the stock prices, calculate some basic metrics, and visualize the data with additional
features like moving averages and visualize the same.'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('stock - Sheet1.csv')
print(df)
print("Stock prices:",df['Stock_Price'])
print("\nBasic Metrics:",df['Stock_Price'].describe())
df['20-day ma']=df['Stock_Price'].rolling(window=20).mean()
df['50-day ma']=df['Stock_Price'].rolling(window=50).mean()
plt.plot(df.index,df['Stock_Price'],label='Stock Price',color='black')
plt.plot(df.index,df['20-day ma'],label='20-day Moving Average')
plt.plot(df.index,df['50-day ma'],label='50-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Simulated Stock Price and Moving Averages (50-day & 200-day)')
plt.legend()
plt.show()
```
